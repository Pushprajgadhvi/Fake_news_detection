```python
import pandas as pd
import numpy as np
```

```python
true = pd.read_csv('True.csv')
true.head()
```

```python
fake = pd.read_csv('Fake.csv')
fake.head()
```

```python
true['label'] = 1
```

```python
fake['label'] = 0
```

```python
news = pd.concat([fake , true], axis=0)
```

```python
news.isnull().sum()
```

```python
news = news.drop(['title', 'subject', 'date'], axis=1)
```

```python
news.head()
```

```python
news = news.sample(frac=1) # it for the shuffling of the data
```

```python
news.head()
```

```python
news = news.reset_index()   #for resetting the index 
```

```python
news.head()
```

```python
news = news.drop(['index'],axis=1)
```

```python
news.head()
```

```python
import re
```

```python
def wordopt(text):
    # Convert into lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove digits
    text = re.sub(r'\d', '', text)

    # Remove newline characters
    text = re.sub(r'\n', ' ', text)

    return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train , y_test = train_test_split(x ,y ,test_size=0.3, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
xv_train = vectorizer.fit_transform(x_train)  # shape (31428, features)
xv_test = vectorizer.transform(x_test) 
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test , y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree  import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(xv_train , y_train)
```

```python
pred_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test, y_test)
```

```python
print(classification_report(y_test , pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train , y_train)
```

```python
predict_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test , y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predict_rfc))
```

```python
def output_label(n):
    if n==0:
        return "It was the fake news"
    elif n == 1:
        return "It was the real news"
```

```python
def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)  # Preprocessing
    new_x_test = new_def_test["text"]
    new_xv_test = vectorizer.transform(new_x_test)  # Vectorize

    pred_lr = LR.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)

    return "\n\nLR Prediction: {} \nRFC Prediction: {}".format(
        output_label(pred_lr[0]),
        output_label(pred_rfc[0])
    )
```
